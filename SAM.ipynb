```python
# Script to rename roboflow filenames to something SAM 2.1 compatible.
# Maybe it is possible to remove this step tweaking sam2/sam2/configs/train.yaml.
import os
import re

FOLDER = "./data/train"

for filename in os.listdir(FOLDER):
    # Replace all except last dot with underscore
    new_filename = filename.replace(".", "_", filename.count(".") - 1)
    if not re.search(r"_\d+\.\w+$", new_filename):
        # Add an int to the end of base name
        new_filename = new_filename.replace(".", "_1.")
    os.rename(os.path.join(FOLDER, filename), os.path.join(FOLDER, new_filename))
```

```python
def predict_sam(image,box,predictor):
    predictor.set_image(image)
    masks, scores, _ = predictor.predict(
        point_coords=None,
        point_labels=None,
        box=box,
        return_logits=True,
        multimask_output=False,
    )
    return masks, scores

import json

with open('../subm.json') as file:
    data = json.load(file)

from tqdm import tqdm
import pycocotools.mask
from PIL import Image

res = []
cnt = 0

for i in tqdm(data):
    IMAGE_PATH = "/kaggle/working/sam2/test/" + i['image_name']
    img = Image.open(IMAGE_PATH)
    bbox = i['bbox']
    bbox[2], bbox[3] = bbox[0] + bbox[2], bbox[1] + bbox[3]
    masks, scores = predict_sam(np.array(img), bbox, predictor)
    if len(list(scores)) > 0:
        if float(scores[0]) <= 0.85:
            res.append(i)
            continue
        cnt += 1
        binary_mask = (masks[0] > 0).astype(np.uint8)
        i['segmentation'] = pycocotools.mask.encode(np.asfortranarray(binary_mask.astype(np.uint8)))
        i['score'] = float(scores[0])
        i['segmentation']['counts'] = str(i['segmentation']['counts'])
        res.append(i)
    else:
        res.append(i)

res1 = []

for i1, i in enumerate(res):
    try:
        i['segmentation']['counts'] = i['segmentation']['counts'].decode("utf-8")
    except:
        res1.append(i)
        continue
    if type(i['segmentation']['counts']) is not str:
        print(123)
    res1.append(i)

import json
import numpy as np

# Convert all numpy float32 values to Python float
def convert_numpy_and_bytes(obj):
    if isinstance(obj, np.float32):  # Конвертация float32 в float
        return float(obj)
    # elif isinstance(obj, bytes):  # Конвертация bytes в строку (utf-8)
    #     return str(obj) #.decode("utf-8")
    elif isinstance(obj, list):  # Рекурсивная обработка списков
        return [convert_numpy_and_bytes(i) for i in obj]
    elif isinstance(obj, dict):  # Рекурсивная обработка словарей
        return {k: convert_numpy_and_bytes(v) for k, v in obj.items()}
    return obj

res2 = convert_numpy_and_bytes(res1)

with open('/kaggle/working/subm.json', "w") as f:
    json.dump(res2, f, indent=4)
```
